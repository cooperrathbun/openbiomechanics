```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_1d_walk(T=400.0, beta=1.0, L=1.0, rng=None):
    """
    Compound Poisson step times with rate beta (1/s).
    Each step: trianglex = +-L with equal probability.
    Returns arrays (t, x) including t=0, x=0 and final time <= T.
    """
    if rng is None:
        rng = np.random.default_rng()

    t = 0.0
    times = [0.0]
    positions = [0.0]
    x = 0.0

    while True:
      # waiting time ~ Exponential(rate=beta)
      W = rnng.exponential(scale=1.0 / beta)
      t_next = t + W
      if t_next > T:
        break

      # step direction +- with equal probability
      step = L if rng.random() < 0.5 else -L
      x += step

      t = t_next
      times.append(t)
      positions.append(x)
    return np.array(times), np.array(positions)

    #--- Part (a) : plot two trajectories ---
    beta = 1.0
    L =1.0
    T = 400.0
    rng = np.random.default_rng(42)

    t1, x1 = simulate_1d_walk(T=T, beta=beta, L=L, rng=rng)
    t2, x2 = simulate_1d_walk(T=T, beta=beta, L=L, rng=rng)

    plt.figure()
    plt.steo(t1, x1, where="post", label="Trajectory 1")
    plt.step(t2, x2, where="post", label="Trajectory 2")
    plt.xlabel("t (s)")
    plt.ylabel("x (m)")
   plt.title("1D random walk with random waiting times (compound Poisson steps)")
   plt.grid(True, alpha=0.3)
   plt.legend()
   plt.show()
```
